```python
import pandas as pd
df = pd.read_csv( "./data/2024_flavors_of_cacoa.tsv", sep='\t')
```

```python
df
```

```python
df.columns
```

```python
df['Cocoa Percent'] = df['Cocoa Percent'].str.rstrip('%').astype(float) / 100
```

```python
df
```

```python
df['Ingredients']
```

```python
ingredients_split = df['Ingredients'].str.split(',', expand=True)
ingredients_split.columns = ['B', 'C', 'L', 'S', 'S*', 'Sa']
df = pd.concat([df, ingredients_split], axis=1)
```

```python
df
```

```python
dummies = df['Most Memorable Characteristics'].str.get_dummies(sep=', ')

char_counts = dummies.sum()

char_counts = char_counts[char_counts >= 20]


dummies = dummies[char_counts.index]

df = pd.concat([df, dummies], axis=1)
```

```python
df
```

```python
df.drop(columns=['Ingredients', 'Most Memorable Characteristics'], inplace=True)
```

```python
df
```

```python
df.to_csv('./data/cleaned_data_2025_flavors_of_cacao.csv', index=False)
```

```python
reduced_df = df[['Review Date', 'Country of Bean Origin','Cocoa Percent', 'Rating']]
reduced_df.to_csv('./data/data_reduced_2025_flavors_of_cacao.csv', index=False)
reduced_df.to_json('./data/data_reduced_2025_flavors_of_cacao.json', orient='records', lines=True)

```

```python
reduced_df
```

```python
filtered_df = df[
    (df['Rating'] >= 3.25) &
    (df['Cocoa Percent'].between(0.65, 0.75)) &
    (df['Review Date'].between(2018, 2021)) &
    ((df["fatty"] == 1) | (df["earthy"] == 1) | (df["roasty"] == 1))
]


filtered_df = filtered_df[['REF', 'Company (Manufacturer)', 'Company Location', 'Review Date',
       'Country of Bean Origin', 'Specific Bean Origin or Bar Name',
       'Cocoa Percent', 'Rating', "fatty", "earthy", "roasty"]]


filtered_df.to_csv('data/data_filtered_2025_flavors_of_cacao.csv', index=False)
filtered_df.to_json('data/data_filtered_2025_flavors_of_cacao.json', orient='records')
```
